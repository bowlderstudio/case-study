# Get max WMXH for specified lat, lon and date

```python
date = "1978-04"
lat = -37.11
lon = 150.0
```

```python
def getMaxWMXH(df, date, lat, lon):
    sub_df = df[date].loc[(df['LATITUDE'] == lat) & (df['LONGITUDE'] == lon)]
    return max(sub_df.WMXH)
```

```python
getMaxWMXH(df, date, lat, lon)
```
